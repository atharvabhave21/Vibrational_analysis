```python
def normalizeDataFrame(df):
    maxReading = df.max(axis = 0)
#     print (maxReading)
    minReading = df.min(axis = 0)
#     print (minReading)
    diff = maxReading - minReading
#     print (diff)
    mean = df.mean(axis = 0)
#     print (mean)
    ndf = (df - mean)/diff
    return ndf

normalizedData = normalizeDataFrame(df)
# print (normalizedData)
```
